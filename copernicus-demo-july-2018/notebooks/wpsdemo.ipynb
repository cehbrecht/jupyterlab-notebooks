## Example with birdy WPS client
https://github.com/bird-house/birdy


```python
from birdy import import_wps as client
```

### Accessing WPS service with CliMAF process 
https://github.com/cp4cds/climaf-wps-demo

```python
climaf = client('https://bovec.dkrz.de/ows/proxy/climaf')
```

### Show available methods/processes

```python
help(climaf)
```

```python
dir(climaf)
```

### Show help for ``tsplot``

```python
help(climaf.tsplot)
```

### Run ``tsplot``

```python
climaf.tsplot()
```

### Show plot

```python
from IPython.display import Image
```

```python
Image(climaf.tsplot())
```

### Use interactive widget

```python
from ipywidgets import interact_manual
```

```python
def f(model, start_year, end_year):
    return Image(climaf.tsplot(model=model, start_year=start_year, end_year=end_year), width=500, height=500)
```

```python
interact_manual(f, 
                model=['MPI-ESM-LR',], 
                start_year=(2010,2100, 10), 
                end_year=(2010, 2100, 10))
```
