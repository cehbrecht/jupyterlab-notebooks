## Example with birdy WPS client
https://github.com/bird-house/birdy


```python
from birdy import import_wps as client
```

### Accessing WPS service with ECAS process 
https://github.com/cehbrecht/ecas-wps-demo

```python
ecas = client('https://bovec.dkrz.de/ows/proxy/ecas')
```

### Show available methods/processes

```python
help(ecas)
```

```python
dir(ecas)
```

### Show help for ``tropical_nights``

```python
help(ecas.tropical_nights)
```

### Run ``tropical_nights``

```python
plot_url = ecas.tropical_nights()
print(plot_url)
```

### Show plot

```python
from IPython.display import Image
```
